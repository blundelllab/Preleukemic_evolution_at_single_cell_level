```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(3, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 9),
        )
        
        def init_weights(m):
            if isinstance(m, nn.Linear):
#                 torch.nn.init.xavier_uniform_(m.weight)
                torch.nn.init.kaiming_uniform_(m.weight)

#                 n = m.in_features
#                 y = 1.0/np.sqrt(n)
#                 m.weight.data.uniform_(-y, y)
#                 m.bias.data.fill_(0)
                
#                 m.weight.data.normal_(0.0,y)
                
        self.linear_relu_stack.apply(init_weights)

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
    
    
```

```python
def test_loop_picking_out_wrong_trees(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    with torch.no_grad():
        for x, y in dataloader:
            num += 16
            pred = model(x)
            for LABEL, tree in zip(y[pred.argmax(1) != y], x[pred.argmax(1) != y]):
                wrong_trees[LABEL] = tree
            for LABEL, tree in zip(y[pred.argmax(1) == y], x[pred.argmax(1) == y]):
                correct_trees[LABEL] = tree
            
    return wrong_trees, correct_trees

def test_loop_picking_out_wrong_trees_ranked_by_uncertainty(dataloader, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    
    for i in range(11):
        wrong_trees[i] = {}

        
    with torch.no_grad():
        for x, y in dataloader:
            for LABEL, tree in zip(y, x):
                
                def convert_format(x):
                    y = []
                    for i in range(16):
                        y.append(x)
                    converted_x = torch.tensor(y) 
                    return converted_x
                
                converted_tree = convert_format(tree.numpy())
                predicted_class_list = predicting_real_tree_params(converted_tree)
                num_of_wrong_guesses = 0
                for guess in predicted_class_list:
                    if guess != LABEL:
                        num_of_wrong_guesses += 1
                num+=1
                if num<20:
                    print('LABEL',LABEL,'predicted_class_list', predicted_class_list)
                    print('num_of_wrong_guesses', num_of_wrong_guesses)
                if num_of_wrong_guesses>0:
                    wrong_trees[num_of_wrong_guesses][LABEL] = tree

    return wrong_trees

def test_loop_picking_out_correct_trees_ranked_by_uncertainty(dataloader, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    
    for i in range(11):
        correct_trees[i] = {}

        
    with torch.no_grad():
        for x, y in dataloader:
            for LABEL, tree in zip(y, x):
                
                
                def convert_format(x):
                    y = []
                    for i in range(16):
                        y.append(x)
                    converted_x = torch.tensor(y) 
                    return converted_x
                
                converted_tree = convert_format(tree.numpy())
                predicted_class_list = predicting_real_tree_params(converted_tree)
                num_of_correct_guesses = 0
                for guess in predicted_class_list:
                    if guess == LABEL:
                        num_of_correct_guesses += 1
                num+=1
                if num<20:
                    print('LABEL',LABEL,'predicted_class_list', predicted_class_list)
                    print('num_of_correct_guesses', num_of_correct_guesses)
                if num_of_correct_guesses>0:
                    correct_trees[num_of_correct_guesses][LABEL] = tree

    return correct_trees
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(4, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 6),
        )
        
        def init_weights(m):
            if isinstance(m, nn.Linear):
#                 torch.nn.init.xavier_uniform_(m.weight)
                torch.nn.init.kaiming_uniform_(m.weight)

#                 n = m.in_features
#                 y = 1.0/np.sqrt(n)
#                 m.weight.data.uniform_(-y, y)
#                 m.bias.data.fill_(0)
                
#                 m.weight.data.normal_(0.0,y)
                
        self.linear_relu_stack.apply(init_weights)

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
    
    
```

```python
def test_loop_picking_out_wrong_trees(dataloader, model, loss_fn, wrong_trees, correct_trees):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    with torch.no_grad():
        for x, y in dataloader:
            num += 16
            pred = model(x)
            for LABEL, tree in zip(y[pred.argmax(1) != y], x[pred.argmax(1) != y]):
                wrong_trees[LABEL] = tree
            for LABEL, tree in zip(y[pred.argmax(1) == y], x[pred.argmax(1) == y]):
                correct_trees[LABEL] = tree
            
    return wrong_trees, correct_trees

def test_loop_picking_out_wrong_trees_ranked_by_uncertainty(dataloader, loss_fn, wrong_trees, correct_trees):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    
    for i in range(11):
        wrong_trees[i] = {}

        
    with torch.no_grad():
        for x, y in dataloader:
            for LABEL, tree in zip(y, x):
                
                def convert_format(x):
                    y = []
                    for i in range(16):
                        y.append(x)
                    converted_x = torch.tensor(y) 
                    return converted_x
                
                converted_tree = convert_format(tree.numpy())
                predicted_class_list = predicting_real_tree_params(converted_tree)
                num_of_wrong_guesses = 0
                for guess in predicted_class_list:
                    if guess != LABEL:
                        num_of_wrong_guesses += 1
                num+=1
                if num<20:
                    print('LABEL',LABEL,'predicted_class_list', predicted_class_list)
                    print('num_of_wrong_guesses', num_of_wrong_guesses)
                if num_of_wrong_guesses>0:
                    wrong_trees[num_of_wrong_guesses][LABEL] = tree

    return wrong_trees

def test_loop_picking_out_correct_trees_ranked_by_uncertainty(dataloader, loss_fn, wrong_trees, correct_trees):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0
    num = 0
    
    for i in range(11):
        correct_trees[i] = {}

        
    with torch.no_grad():
        for x, y in dataloader:
            for LABEL, tree in zip(y, x):
                
                
                def convert_format(x):
                    y = []
                    for i in range(16):
                        y.append(x)
                    converted_x = torch.tensor(y) 
                    return converted_x
                
                converted_tree = convert_format(tree.numpy())
                predicted_class_list = predicting_real_tree_params(converted_tree)
                num_of_correct_guesses = 0
                for guess in predicted_class_list:
                    if guess == LABEL:
                        num_of_correct_guesses += 1
                num+=1
                if num<20:
                    print('LABEL',LABEL,'predicted_class_list', predicted_class_list)
                    print('num_of_correct_guesses', num_of_correct_guesses)
                if num_of_correct_guesses>0:
                    correct_trees[num_of_correct_guesses][LABEL] = tree

    return correct_trees
```
